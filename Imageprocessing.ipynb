```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Replace with your image path
image = cv2.imread(r"/content/rafale.png")

#convert BGR (OpenCV) to RGB (Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#display image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

```

```python
resized = cv2.resize(gray, (500, 300))
plt.imshow(resized)

plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
#Blur an image
blurred = cv2.GaussianBlur(resized, (15, 15), 0)

plt.imshow(blurred)
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

```python
#allows us to see complete boundaries of the image
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
plt.show()
```

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Binary Threshold")
plt.axis('off')
plt.show()
#
```

```python
height, width = image_rgb.shape[:2]
center = (width / 2, height / 2)

matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated_image = cv2.warpAffine(image_rgb, matrix, (width, height))

plt.imshow(rotated_image)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
cv2.imwrite("edges_output.jpg", edges)
print("Image Saved successfully")

```

```python
fig, axes = plt.subplots(2,2, figsize=(10,8))

axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Original")
axes[0, 0].axis("off")

axes[0,1].imshow(gray, cmap='gray')
axes[0,1].set_title("Grayscale")
axes[0,1].axis("off")

axes[1,0].imshow(blurred)
axes[1,0].set_title("Blurred")
axes[1,0].axis("off")

axes[1,1].imshow(edges, cmap='gray')
axes[1,1].set_title("Edges")
axes[1,1].axis("off")

plt.tight_layout()
plt.show()
```
